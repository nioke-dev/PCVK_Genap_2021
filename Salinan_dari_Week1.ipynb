```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

# Membuat List untuk menyimpan URL dari beberapa image
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://d1vbn70lmn1nqe.cloudfront.net/prod/wp-content/uploads/2021/10/26071254/mengenal-fakta-menarik-seputar-kucing-anggora-turki-halodoc.jpg"
]

# Baca dan tampilkan image
# loop pada setiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url) # Membaca gambar dari URL
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5) # Ubah ukuran gambar menjadi setengahnya
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) # Mengubah warna gambar dari BGR ke RGB
    final_frame = cv.hconcat((image, image_2)) # Menggabungkan gambar asli dan gambar RGB secara horizontal
    cv2_imshow(final_frame) # Menampilkan gambar
    print('\n')
```
