## 3. Kegunaan Kode Program pada Langkah ke-9 dan Pengaruhnya Jika Tidak Dilakukan
Potongan kode program berikut:
```python
image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
```
Apa kegunaan kode program tersebut?dan apa pengaruhnya jika tidak dilakukan?

*   **Kegunaan:** Kode ini digunakan untuk mengubah ukuran gambar menjadi setengah dari ukuran aslinya. `fx=0.5` dan `fy=0.5` menunjukkan faktor pengali untuk lebar dan tinggi gambar, sehingga ukuran gambar akan dikurangi menjadi 50% dari ukuran aslinya.
*   **Pengaruh Jika Tidak Dilakukan:** Jika langkah ini tidak dilakukan, gambar akan tetap pada ukuran aslinya. Hal ini mungkin menyebabkan gambar terlalu besar untuk diproses atau ditampilkan dengan benar, terutama jika bekerja dengan gambar beresolusi tinggi.

## 4. Perhatikan potongan kode progam berikut :
```python
# membuat garis horizontal ditengah image
for y in range (lebar) :
  image_3[int((tinggi)/2),y] = [255,255,255]
```
Apakah kegunaan kode `[255,255,255]` ? Jelaskan!


*   **Kegunaan:** `image_3[int((tinggi)/2), y] = [255, 255, 255]` digunakan untuk membuat garis horizontal putih di tengah gambar. `[255, 255, 255]` adalah representasi warna putih dalam format RGB (Red, Green, Blue), di mana setiap nilai 255 menunjukkan intensitas maksimum dari masing-masing kanal warna.
*   **Penjelasan:** Kode ini menetapkan pixel pada posisi tertentu menjadi putih. Dengan melakukan loop melalui lebar gambar (`range(lebar)`), seluruh baris di tengah gambar diubah menjadi warna putih, menciptakan garis horizontal.

## 5. Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!


*   Pixel: Pixel (picture element) adalah unit terkecil dari gambar digital yang memiliki informasi warna. Setiap gambar terdiri dari sejumlah pixel yang membentuk keseluruhan gambar.
*   Resolusi Gambar: Resolusi gambar mengacu pada jumlah total pixel dalam gambar, biasanya dinyatakan sebagai lebar x tinggi (misalnya, 1920x1080). Resolusi yang lebih tinggi berarti lebih banyak pixel yang membentuk gambar.
*   Keterkaitan: Resolusi yang tinggi menghasilkan gambar yang lebih detail karena ada lebih banyak pixel yang merepresentasikan setiap bagian gambar. Sebaliknya, resolusi rendah mengakibatkan gambar terlihat lebih buram atau pecah karena jumlah pixel yang lebih sedikit. Resolusi tinggi diperlukan untuk menampilkan detail yang lebih jelas dalam gambar, sedangkan resolusi rendah sering kali digunakan untuk menghemat ruang penyimpanan atau bandwidth.

````python
import cv2 as cv
import numpy as np
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

# URL gambar KTP
url = "https://th.bing.com/th/id/OIP.DW05DkkCw1b0lDw0rCg9sQAAAA?rs=1&pid=ImgDetMain"

# Mengunduh gambar dari URL
response = requests.get(url)
image_data = BytesIO(response.content)

# Membaca gambar dengan OpenCV
image = np.array(Image.open(image_data))
image = cv.cvtColor(image, cv.COLOR_RGB2BGR)  # Konversi ke format BGR untuk OpenCV

# Konversi gambar ke format RGB untuk matplotlib
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan koordinat area NIK (x, y, lebar, tinggi)
# Anda perlu mengganti nilai ini dengan koordinat yang sesuai
x, y, lebar, tinggi = 148, 98, 153, 15  # Ganti dengan koordinat yang sesuai

# Menyensor area NIK dengan warna hitam
image_rgb[y:y+tinggi, x:x+lebar] = [255, 255, 255]

# Menampilkan gambar dengan matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis('on')  # Matikan axis untuk tampilan gambar yang bersih
plt.show()
````
