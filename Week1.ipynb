### 4. Kegunaan Kode `[255, 255, 255]`
Kode `[255, 255, 255]` biasanya mewakili warna **putih** dalam representasi warna RGB. Setiap angka pada array tersebut menunjukkan intensitas warna pada masing-masing kanal (Red, Green, dan Blue) dengan skala dari 0 hingga 255.
- **Kegunaan**: Kode ini digunakan untuk mengatur warna tertentu, misalnya untuk membuat latar belakang putih, menggambar teks putih di atas gambar, atau menandai pixel putih saat melakukan manipulasi gambar.
- **Contoh**: Saat menggambar teks di gambar dengan warna putih menggunakan OpenCV:
    ```python
    cv.putText(image, 'Hello', (50, 50), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    ```
    Di sini, `(255, 255, 255)` menetapkan bahwa teks akan berwarna putih.

### 5. Keterkaitan antara Pixel dan Resolusi Gambar
- **Pixel**: Pixel (picture element) adalah unit terkecil dari gambar digital. Setiap pixel memiliki informasi warna yang berkontribusi pada keseluruhan gambar.
- **Resolusi Gambar**: Resolusi gambar mengacu pada jumlah total pixel yang ada dalam sebuah gambar, biasanya dinyatakan dalam bentuk lebar x tinggi (misalnya, 1920x1080).
- **Keterkaitan**: Resolusi yang lebih tinggi berarti lebih banyak pixel yang membentuk gambar. Ini menghasilkan detail yang lebih tinggi karena setiap bagian kecil dari gambar memiliki lebih banyak informasi warna. Gambar dengan resolusi rendah akan terlihat lebih buram atau berbintik karena memiliki lebih sedikit pixel untuk merepresentasikan detail yang sama.
